```python
import tkinter as tk
from tkinter import messagebox

class Calculator:
  @staticmethod
  def add_numbers(a, b):
    return f'Addition: {a + b}'

  @staticmethod
  def subtract_numbers(a, b):
    return  f'Subtraction: {a - b}'

  @staticmethod
  def multiply_numbers(a, b):
    return f'Multiplication: {a * b}'

  @staticmethod
  def divide_numbers(a, b):
        if b == 0:
            return "Error!"
        return f"Division: {a / b}"
      
  @staticmethod
  def percentage(a, b):
      if b == 0:
          return "Error!"
      return f"Percentage: {(a * 100) / b}"
def Calculate():
    try:
        a = int(entry1.get())
        b = int(entry2.get())
        operator = operator_var.get()
        
        if operator == "+":
            result = Calculator.add_numbers(a, b)
        elif operator == "-":
            result = Calculator.subtract_numbers(a, b)
        elif operator == "*":
            result = Calculator.multiply_numbers(a, b)
        elif operator == "/":
            result = Calculator.divide_numbers(a, b)
        elif operator == "%":
            result = Calculator.percentage(a, b)
        else:
            result = "Invalid Operator"

        result_label.config(text=result)
    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter valid numbers")

window = tk.Tk()
window.title("Simple Calculator")

tk.Label(window, text="Enter first number:").grid(row=0, column=0)
entry1 = tk.Entry(window)
entry1.grid(row=0, column=1)

tk.Label(window, text="Enter second number:").grid(row=1, column=0)
entry2 = tk.Entry(window)
entry2.grid(row=1, column=1)

tk.Label(window, text="Choose operator:").grid(row=2, column=0)
operator_var = tk.StringVar()
operator_menu = tk.OptionMenu(window, operator_var, "+", "-", "*", "/", "%")
operator_menu.grid(row=2, column=1)

tk.Button(window, text="Calculate", command=calculate).grid(row=3, column=0, columnspan=2)

result_label = tk.Label(window, text="Result will appear here")
result_label.grid(row=4, column=0, columnspan=2)

window.mainloop()

```
